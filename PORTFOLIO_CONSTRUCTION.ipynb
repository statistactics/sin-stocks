# Reading in Files

```python
stocks_df = pd.read_csv("final_merged.csv", parse_dates = ["Unnamed: 0"])
stocks_df.set_index("Unnamed: 0", inplace=True)
stocks_df.index.name = "Date"
stocks_df.head()
```

# Further Data Cleaning

## Isolating Sin Stocks Only

```python
# Removing the NaN value in sin_list
sin_list = [x for x in sin_list if x != "nan"]

sin_list
```

```python
sin_stocks = stocks_df[sin_list]
sin_stocks.head()
```

## Finding Percentage Change in Stocks

```python
returns_df = sin_stocks.pct_change(limit = None, fill_method = None)
returns_df
```

Using this set of arguments allows for the pct change values to be properly reflected when the stock exits the universe

### Removing null values
- Only the first row is empty, since we found percentage changes for each adjacent row

```python
returns_df.dropna(how = "all", inplace=True)
# returns_df.to_csv("price_returns.csv")
```

## Calculating Individual Weights

```python
total_no_sin = len(returns_df.columns)
total_no_sin
```

### Calculating Number of Active Stocks for Each Date
- Deducting number of NaN values per row from the total number of sin stocks to derive the number of active sin stocks for a given date

```python
returns_df["Num_Active_Stocks"] = total_no_sin - returns_df.isnull().sum(axis = 1)
returns_df["Portfolio_Proportion"] = 1 / returns_df["Num_Active_Stocks"]
returns_df
```

## Multiplying percentage returns by 1/N stocks (Portfolio_Proportion)

```python
portfolio_returns = returns_df.mul(returns_df["Portfolio_Proportion"], axis = 0)
portfolio_returns.drop(columns = ["Num_Active_Stocks", "Portfolio_Proportion"], inplace = True)
```

## Finding Row Wise Sum of Portfolio Percentage Returns

```python
portfolio_returns["Portfolio_Returns"] = portfolio_returns.sum(axis = 1)
```

```python
portfolio_returns["Portfolio_Proportion"] = returns_df["Portfolio_Proportion"]
portfolio_returns
```

```python
portfolio_merged = pd.concat([returns_df[["Num_Active_Stocks"]], portfolio_returns], axis = 1, join = "inner")
```
